```python
import numpy as np
import pandas as pd
import cv2
import pytesseract
import os
from glob import glob
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
import easyocr
```

```python
imagePaths= glob('./invoices/*.*')
```

```python
imagePaths
```

```python
allInvoices = pd.DataFrame(columns=['id','text'])
for imagePath in tqdm(imagePaths, desc='Invoice'):
    #imagePath = imagePaths[0]
    _, filename = os.path.split(imagePath)
    print(filename)
    # extract data and text
    image= cv2.imread(imagePath)
    reader = easyocr.Reader(['en'], gpu=True)
    result = reader.readtext(image)
    print(result)
    df = pd.DataFrame(columns=['text','conf','id'])
    for (bbox, text, prob) in result: 
#     print(text, prob)
        invoiceData = pd.DataFrame()
        invoiceData['text'] = [text]   
        invoiceData['conf'] = [prob]
        invoiceData['id'] = [filename]
        df = pd.concat((df, invoiceData))

#     df['conf'] = df['conf'].astype(float)
#     useFulData = df.query('conf >=30')
    # Dataframe 
#     invoice = pd.DataFrame()
#     invoice['text'] = df['text'].astype(String)
#     invoice['id'] = [filename]
    # Concatenation 
    allInvoices = pd.concat((allInvoices, df))
```
